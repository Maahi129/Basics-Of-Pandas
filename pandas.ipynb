```python
#Renaming

Demo.rename(
                                            # we can rename columns with rename keyword
    columns ={
        "Sex" : "Gender"
    }

).head(1)
```

```python
# Aggregating Statistics

Demo.agg(                                # Determining mean,median,min and max with agg|
{
    "Age" : ["min","max","mean","median","skew"],
    "Fare": ["min", "max", "mean","skew","median"],
})
```

```python
# Groupby

print(Demo[["Sex","Age"]].groupby("Sex").mean())   #Grouping Data According to categories
d = Demo.groupby("Embarked")                       #grouping Embarked and Printing First entries
d.first()
```

```python
# Mean , Medain 

display(Demo.groupby("Sex")["Age"].median())   # Grouping Sex And finding Age median
display(Demo.groupby(["Sex","Parch"])["Fare"].mean())     # Grouping Sex , Parch And Finding Fare mean 
```

```python
# Value Counts

Demo["Age"].value_counts().head(5)        # value counts counts number of records in each category in a cloumn
```

## CLEANING DATA

### Data cleaning means fixing bad data in your data set.

* Bad data could be:
* 1.Empty Cells
* 2.Data in Wrong format
* 3.Wrong Data
* 4.Duplicates

```python
# Empty Cells
# We can remove rows with dropna which having null values

#Demo["Cabin"].dropna()
```

```python
# With Fillna we can Fill null values with some values (Used Rather Than Dropna)
Demo1 = pd.DataFrame(
   {
       "Age" : [18, None],
       "gender": ["male", "female"],
       "Loaction" : ["India", "Swiss"],
   }
)
Demo1["Age"].fillna(value = 65, inplace=True)     # Filling null values with the help of nullna
Demo1
```

```python
# Replacing values

Demo1.loc[1 , "Age"] = 20             # Replacing 65 with 20 
Demo1                                 
```

```python
# Duplicates
Demo.duplicated()      #  If they Have same Values in all set values it shows True  other wise false (It's not duplicated)
```

```python
Demo = pd.read_csv("Documents/DemoList.csv")
Demo.sort_values("Name", axis=0, ascending = True, inplace = True)  # Sorting NAme in Alphabetical Order
Demo
```

```python
# Co-relations

Demo.corr()

# corr Show the relationship between the columns
# The corr() method ignores "not numeric" columns.
# The number varies from -1 to 1.
# 1 means that there is a 1 to 1 relationship (a perfect correlation)
# 0.9 is also a good relationship, and if you increase one value, the other will probably increase as well.
# -0.9 would be just as good relationship as 0.9
# 0.2 means NOT a good relationship
```

```python
# Manipulating Textual Data

print(Demo['Name'].str.lower().head(3))    # Lowering Names
print('----------------------------------------')
print(Demo['Name'].str.split(",").str.get(0).head(3)) #Splitting Names and printing starting name
```
